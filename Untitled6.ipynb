# Working with MongoDB

import necessary packages

```python
import pandas as pd
from pymongo import MongoClient
```

```python
#create a connection to the MongoDB
client=MongoClient('localhost',27017)
```

```python
#set the database and the connection
db=client.pythonbicookbook
collection=db.accidents
```

```python
#find the 6th day of the week data
data=collection.find({'Day_of_Week':6})
data
```

```python
# Create the dataframe from the MongoDB query and show the top 5 rows
accidents = pd.DataFrame(list(data))
accidents.head()
```

# Creating dataframe from a csv file

```python
import pandas as pd

accidents_data_file='accidents.csv'
accidents=pd.read_csv(accidents_data_file,
                     sep=',',
                     index_col=0,
                     header=0,
                     parse_dates=True,
                     tupleize_cols=False,
                     error_bad_lines=False,
                     warn_bad_lines=True,
                     skip_blank_lines=True)
```

```python
accidents.head()
```
