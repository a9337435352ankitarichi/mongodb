# Generating summary statistics for the entire dataset

```python
import pandas as pd

accidents_data_file='accidents.csv'

accidents=pd.read_csv(accidents_data_file,
                     sep=',',
                     index_col=False,
                     header=0,
                     parse_dates=['Date'],
                     dayfirst=True,
                     tupleize_cols=False,
                     error_bad_lines=True,
                     warn_bad_lines=True,
                     skip_blank_lines=True)
```

Generating summary statistics of the entire accidents dataset

```python
accidents.describe()
```

To make the summary statistics more readable we transpose the describe results

```python
accidents.describe().transpose()
```

# Getting a count of unique values for a single column

```python
import pandas as pd

accidents_data_file='accidents.csv'

accidents=pd.read_csv(accidents_data_file,
                     sep=',',
                     header=0,
                     index_col=False,
                     tupleize_cols=False,
                     parse_dates=['Date'],
                     dayfirst=True,
                     error_bad_lines=True,
                     warn_bad_lines=True,
                     skip_blank_lines=True)
```

```python
#to see unique value counts for
pd.value_counts(accidents['Date'])
```

# Getting the minimum and maximum values of a single column

```python
# Get the count of each unique value in the Date column.
print("Min Value: {}".format(accidents['Number_of_Vehicles'].min()))
print("Max Value: {}".format(accidents['Number_of_Vehicles'].max()))
```

# Generating quantiles for a single column

```python
#creating quantiles for the column Number_of_Vehicles
accidents['Number_of_Vehicles'].quantile([.05, .1, .25, .5, .75, .9, .99])
```

# Getting the mean,median,mode and range of a single column

```python
print("Mean: {}".format(accidents['Number_of_Vehicles'].mean()))
```

```python
print("Median: {}".format(accidents['Number_of_Vehicles'].median()))
```

```python
print("Mode: {}".format(accidents['Number_of_Vehicles'].mode()))
```

```python
print("Range: {}".format(range(accidents['Number_of_Vehicles'].min(),accidents['Number_of_Vehicles'].max())))
```
